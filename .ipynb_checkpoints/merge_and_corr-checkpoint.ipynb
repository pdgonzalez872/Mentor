```python
import pandas as pd
import os
import matplotlib
%pylab inline
```

```python
# import all data
atp_data = pd.read_csv("data/05262015_Rankings_ATP.csv", delimiter="\t")
country_symbols = pd.read_csv("data/country_symbols.txt", delimiter="\t", header=None, names=["CountryID", "CountryName"])
good_country_index = pd.read_csv("data/good_country_index.txt")
country_transparency = pd.read_csv("data/country_transparency_index.txt")
```

```python
print(atp_data.head())
print(country_symbols.head())
print(good_country_index.head())
print(country_transparency.head())
```

```python
atp_data_country = pd.merge(atp_data, country_symbols, left_on="Country", right_on="CountryID", how="left")
atp_data_country
```

```python
mystery_country_codes = pd.Series(atp_data_country[pd.isnull(atp_data_country.CountryName)]["Country"].unique())

mystery_country_codes.to_csv("data/mystery_countries.csv")
```

```python
mystery_country_answers = pd.read_csv("data/mystery_countries_answers.csv", header=None, names=["CountryID", "CountryName"])
mystery_country_answers
```

```python
atp_data_country = pd.merge(atp_data_country, mystery_country_answers, 
                            left_on="Country", right_on="CountryID", how="left")
```

```python
atp_data_country["Country"] = atp_data_country["CountryName_x"].fillna(atp_data_country["CountryName_y"])
atp_data_country.head()

# need to replace some names so we can merge
atp_data_country.ix[atp_data_country["Country"]=='Russian Federation', "Country"] = "Russia"
atp_data_country.ix[atp_data_country["Country"]=='Korea, Republic of', "Country"] = "South Korea"
atp_data_country.ix[atp_data_country["Country"]=='Moldova, Republic of', "Country"] = "Moldova"
atp_data_country.ix[atp_data_country["Country"]=='Venezuela, Bolivarian Republic of', "Country"] = "Venezuela"
atp_data_country.ix[atp_data_country["Country"]=='Macedonia, the former Yugoslav Republic of', "Country"] = "Macedonia"
atp_data_country.ix[atp_data_country["Country"]=='Philipines', "Country"] = "Philippines"
atp_data_country.ix[atp_data_country["Country"]=='Bolivia, Plurinational State of', "Country"] = "Bolivia"
```

```python
good_country_index.head()
```

```python
atp_good_country = pd.merge(atp_data_country, good_country_index.rename(columns={"Country":"GoodCountryName"}),
                            left_on="Country", right_on="GoodCountryName", how="inner")
atp_good_country[pd.isnull(atp_good_country.GoodCountryName)].Country.unique()
```

```python
atp_good_country = atp_good_country.sort("Ranking").rename(columns={"Rank":"CountryRank"})
atp_good_country
```

```python
pd.median()
```

```python
to_correlation_df = atp_good_country.groupby("Country")[["Career_Prize_Money", "Science and Technology", 
                                                         "Culture",	"International Peace and Security",
                                                         "World Order",	"Planet and Climate",
                                                         "Prosperity and Equality",	"Health and Well being"]].aggregate(median)
to_correlation_df
```

```python
to_correlation_df.plot(x="Career_Prize_Money",y="International Peace and Security", kind="scatter")
```

```python
to_correlation_df.corr()
```
